```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\Toshiba\Desktop\WHO.csv")
dataset.head(20)
```

```python
df = dataset
df['day'] = pd.DatetimeIndex(df['date']).day
data = df.drop(['date'], axis = 1)
data.head(20)
```

```python
data.new_deaths = data.new_deaths.fillna((data.new_deaths.shift() + data.new_deaths.shift(-1))/2)
data.new_cases = data.new_cases.fillna((data.new_cases.shift() + data.new_cases.shift(-1))/2)
data.total_cases = data.total_cases.fillna((data.total_cases.shift() + data.total_cases.shift(-1))/2)
data.total_deaths = data.total_deaths.fillna((data.total_deaths.shift() + data.total_deaths.shift(-1))/2)
```
