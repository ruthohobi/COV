```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
dataset = pd.read_csv(r"C:\Users\Toshiba\Desktop\WHO.csv")
dataset.head(20)
```

```python
dataset.dropna(inplace=True)
```

```python
print(dataset)
```

```python
dataset.shape
```

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'new_cases', color = 'maroon', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='blue', markersize=6, figsize = (14,7))
plt.title(' Total Number of New Cases in USA and Canada - fig 1.1')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'new_deaths', color = 'blue', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='maroon', markersize=6, figsize = (14,7))
plt.title(' Number of New Deaths in USA and Canada - fig 1.2')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.legend()
plt.show()
```

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'total_cases', color = 'red', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='green', markersize=6, figsize = (14,7))
plt.title(' Total Number of Cases in USA and Canada - fig 1.3')
plt.xlabel('Date')
plt.ylabel('Total Cases')
plt.legend()
plt.show()
```

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'total_deaths', color = 'green', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='red', markersize=6, figsize = (14,7))
plt.title(' Total Number of Deaths in USA and Canada - fig 1.4')
plt.xlabel('Date')
plt.ylabel('Total Deaths')
plt.legend()
plt.show()
```
