```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
dataset = pd.read_csv(r"C:\Users\Toshiba\Desktop\WHO.csv")
dataset.head(20)
```

```python
dataset.dropna(inplace=True)
```

```python
print(dataset)
```

```python
dataset.shape
```

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'new_cases', color = 'maroon', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='blue', markersize=6, figsize = (14,7))
plt.title(' Total Number of New Cases in Canada and US - fig 1.1')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.show()
```

#### From the above plot, it is seen that there are very few new cases of COVID in Canada and it is on a steady rate. However, it is not the same for US, which has had a rapid increase in the number of new cases. This could be as a result of so many factors, one of the major which is the US astronomical population.

```python
plt.clf()
df = dataset[dataset['location'].isin(['Canada' ,'United States'])]
df.plot(x = 'date', y = 'new_deaths', color = 'blue', kind = 'line', linewidth = 3, marker = 'o', 
        markerfacecolor='maroon', markersize=6, figsize = (14,7))
plt.title(' Number of New Deaths in Canada and US - fig 1.2')
plt.xlabel('Date')
plt.ylabel('New Deaths')
plt.legend()
plt.show()
```

#### From the above plot we can see that Canada had a steady death rate but as the day progresses, there were new recorded deaths. This can be as a result of people not calling the emergency line when they show possible symptoms of COVID, so they can be treated early. The same thing is applicable to the US which have a very high death rate.
